```python
weather_imp['target'] = weather_imp.shift(-1)['temp_max']
weather_imp
```

```python
from sklearn.linear_model import Ridge
#this type minimize overfitting
reg = Ridge(alpha=.1)
```

```python
predictors = ['precip', 'temp_max', 'temp_min']
```

```python
train_set = weather_imp.loc[:'2015-01-01']
test_set = weather_imp.loc['2015-01-01':]
print(train_set.shape , test_set.shape)
```

```python
reg.fit(train_set[predictors], train_set['target'])
```

```python
predictions = reg.predict(test_set[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(test_set['target'], predictions)
```

```python
combine = pd.concat([test_set['target'], pd.Series(predictions, index= test_set.index)], axis=1)
combine.columns = ['actuals', 'predictions']
combine
```

```python
def predict(predictors, weather_imp, reg):
    train_set = weather_imp.loc[:'2015-01-01']
    test_set = weather_imp.loc['2015-01-01':]
    reg.fit(train_set[predictors], train_set['target'])
    predictions = reg.predict(test_set[predictors])
    error = mean_absolute_error(test_set['target'], predictions)
    combine = pd.concat([test_set['target'], pd.Series(predictions, index= test_set.index)], axis=1)
    return error, combine
```
